```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("day.csv")
```

```python
df
```

```python
labels = ['January 2011','February 2011','March 2011', 
          'April 2011','May 2011']
fig, ax = plt.subplots(figsize=(15, 8))
ax.plot(df['dteday'],df['casual'], color = 'red', label = 'casual')
ax.plot(df['dteday'],df['registered'], color = 'green',alpha =0.5, 
        label = 'registered')
ax.set_xticklabels(labels)
ax.legend()
plt.show()
```

```python
df_2011 = df[df['yr']==0]
casual_2011 = df_2011.groupby('mnth')['casual'].mean()

fig, ax=plt.subplots(figsize=(10,8))
ax.plot(casual_2011.index, casual_2011.values)
```
